# Generate Author Statistics

This notebook groups the article dataset by author and computes summary statistics (e.g. mean/median/count/sum), which are joined onto the article story features dataset (df_story_features_chunk1-4.csv)

```python
import numpy as np
import pandas as pd
from IPython.core.display import display, HTML
```

```python
data = pd.read_csv("data/df_story_noHTML_fixCols.csv")
```

```python
data = data[-data['PostID'].isnull()]      

# Filter NaN ResponseTime
data = data[-data['ReadingTime'].isnull()] 

# Filter stories posted after April 2020
data['PublishedDate'] = pd.to_datetime(data['PublishedDate']).dt.date
data = data[data['PublishedDate'] < pd.to_datetime("2020-04-01")]
```

```python
data.columns
```

```python
authorStats = data[['ClapCount_Story', "ResponseNum_Story", "VoterCount", "isPaywall"]].groupby(data["User"]).agg(['mean', 'median', 'count', 'sum'])
```

# There are 1,150,786 unique authors

```python
print(len(authorStats))
display(authorStats.head())
authorStats.columns.set_levels(['AuthorArticlesClap', 'AuthorArticlesResponse', 'AuthorArticlesVoter', 'AuthorArticlesisPaywall'],level=0,inplace=True)
authorStats.columns = ['_'.join(col).strip() for col in authorStats.columns.values]
# authorStats.iloc[300]
```

```python
authorStats.head()
```

```python
authorStats.to_csv("data/clean/authorStats.csv", header=True, index=True)
```
