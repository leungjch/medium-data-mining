```python
import numpy as np
import pandas as pd
from IPython.core.display import display, HTML
```

```python
df = pd.read_csv("data/df_story_noHTML_clean.csv")
```

```python
df.columns
```

```python
tagStats = df[['ClapCount_Story', "ResponseNum_Story", "ReadingTime", "VoterCount", "isPaywall"]].groupby(df["TagSource"]).agg(['mean', 'median', 'count', 'sum', 'std'])
```

```python
tagStats.head(1)
```

```python
tagStats.columns.set_levels(['TagClap', 'TagResponse', 'TagReadingTime', 'TagVoterCount', 'TagisPaywall'],level=0,inplace=True)
tagStats.columns = ['_'.join(col).strip() for col in tagStats.columns.values]
```

```python
tagStats.sample(10)
```

```python
tagStats.to_csv("data/clean/tagStats.csv", index=True, header=True)
```
