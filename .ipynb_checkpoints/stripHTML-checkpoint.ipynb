```python
import pandas as pd
count = 0
header = True
for data in pd.read_csv('df_story_chunk3.csv',encoding = 'ISO-8859-1',chunksize  = 100000, low_memory=False):
    count += 1                          # counting the number of chunks
    lastlen = len(data)                 # finding the length of last chunk
    data['StoryHTML'] = ''
    data.to_csv("df_story_chunk3_noHTML.csv", header=False, mode="a")
    header = False
datalength = (count*100000 + lastlen - 100000) # length of total file
print(datalength)
```
